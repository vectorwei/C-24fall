```python
import matplotlib.pyplot as plt
import numpy as np
# 从文本文件中读取数据
with open('C_times.txt', 'r') as file:
    lines = file.readlines()
sizes_A = []
times1 = []
for line in lines[1:]:
    data = line.strip().split('\t')
    sizes_A.append(data[0])
    times1.append(float((data[2])))


with open('java_times.txt', 'r') as file:
    lines = file.readlines()
times2 = []
for line in lines[1:]:
    data = line.strip().split('\t')
    times2.append(float((data[2])))

# 绘制折线图
plt.plot(sizes_A, np.log(times1), label='C running Time')
plt.plot(sizes_A, np.log(times2), label='java running Time')
# 设置图表标题和标签
plt.title('Matrix Multiplication Time')
plt.xlabel('Matrix Size')
plt.ylabel('log-Time (seconds)')

# 添加图例
plt.legend()

# 旋转 x 轴标签
plt.xticks(rotation=45, ha='right')

# 显示网格
plt.grid(True)

# 显示图表
plt.show()
```
